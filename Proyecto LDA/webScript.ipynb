```python
import requests
from bs4 import BeautifulSoup

def obtenerLinks(link): #Se da un link 'https://presidente.gob.mx/sala-de-prensa/boletines/'
    req = requests.Session() #Request
    response = req.get(link, headers={'Accept': 'application/xml; charset=utf-8','User-Agent':'foo'}) #Guardar la respuesta con usuario dinamico
    coverpage = response.content #Guardar contenido
    soup = BeautifulSoup(coverpage, 'html.parser') #Convertir contenido a html
    URLs = [] #Arreglo para guardar los enlances a las noticias que contiene la pagina
    for i in soup.find_all("div", attrs={"class" : "entry-media"}):
        for j in i.find_all("h3"):
            for k in j.find_all('a',href=True):
                URLs.append(k['href'])
    return URLs

```

```python
links = [] #Aqui se guardaran todos los links obtenidos
nuevosLinks = [] #temporal para guardar links de cada iteracion
for i  in range (1, 315):
    nuevosLinks = obtenerLinks(f"https://presidente.gob.mx/sala-de-prensa/boletines/page/{i}/")
    links.extend(nuevosLinks)
    print("Pagina" + str(i))
print(links)
```

```python
f = open("Links.txt", "x") # Escribir en un archivo el texto html
for x in range(len(links)):
    f.write(str("\"" + links[x] + str("\"")+ str(",")))
f.close()
```

```python
import requests
import re
import os #para checar si existen nombres del mismo archivo
def descargarHTML(links):
    for i in range(len(links)):
        req = requests.Session() #Request
        response = req.get(links[i], headers={'Accept': 'application/xml; charset=utf-8','User-Agent':'foo'}) #Guardar la respuesta con usuario dinamico
        coverpage = response.content #Guardar contenido
        soup = BeautifulSoup(coverpage, 'html.parser') #Convertir contenido a html

        date = re.findall("([0-9]{2}\/[0-9]{2}\/[0-9]{4})", str(soup.find("span"))) #Expresion regular para encontrar la fecha en el html
        date[0] = date[0].replace("/", "-") #Reemplazar las diagonales de la fecha por guiones para usarlo en el nombre del archivo
        counter = 1 #contador para archivos con mismo nombre, se reinicia a 0 al final de cada iteracion del for
        filename = f"boletines/{date[0]}.html"
        while os.path.exists(filename): #mientras siga existiendo archivos con el mismo nombre se le agregan numeros crecientes
            filename = f"boletines/{date[0]}({counter}).html"
            counter = counter + 1 #1,2,3... hasta que ya no existen archivos con la misma fecha
        counter = 1 #reiniciar el contador de las fechas
        f = open(filename, "x") # Escribir en un archivo el texto html
        print(f.write(str(soup)))
        f.close()
```

```python
descargarHTML(links)
```
