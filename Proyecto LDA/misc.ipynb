```python
import os
import re
import shutil
```

```python
file_list = os.listdir("./data2")
file_list
```

```python
def reverse_date(file_name):
    # Split file name into date components and extension
    date_part, ext = file_name.split('.')
    day, month, year = re.split(r'[-(]', date_part)[:3]

    # Reassemble the file name in the "yyyy-mm-dd" format
    reversed_date = f'{year}-{month}-{day}'
    remaining_parts = file_name[len(reversed_date):]
    return f'{reversed_date}{remaining_parts}'

# Apply the reverse_date function to all file names in the list
reversed_dates = [reverse_date(file_name) for file_name in file_list]

reversed_dates
```

```python
for original_file, reversed_file in zip(file_list, reversed_dates):
    print(f"{original_file}, {reversed_file}")
```

```python
# Copy the content of the original files to the new files in the "reversed" directory,
# or create an empty file if the source file is not found
for original_file, reversed_file in zip(file_list, reversed_dates):

    target_path = os.path.join("data", reversed_file)
    source_path = os.path.join("data2", original_file)

    if os.path.exists(source_path):
        shutil.copyfile(source_path, target_path)
    else:
        print("No data")
        break

print("Files copied or created successfully in the '/reversed' directory.")
```
